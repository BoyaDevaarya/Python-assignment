```python
# Program to check if a number is a palindrome

def isPalindrome(x):
    # convert number to string
    s = str(x)
    
    # check if string is same as reversed string
    if s == s[::-1]:
        return True
    else:
        return False

# Example 1
x = 121
print("Example 1 Input:", x)
print("Output:", isPalindrome(x))  # True
print()

# Example 2
x = -121
print("Example 2 Input:", x)
print("Output:", isPalindrome(x))  # False

```
